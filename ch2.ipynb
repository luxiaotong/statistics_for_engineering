```python
# 2.1
def pareto_plot(x, y):
    df = pd.DataFrame({
        'x': x,
        'y': y
    })
    df = df.sort_values(by='y', ascending=False)
    
    fig, ax1 = plt.subplots()
    ax1.bar(df['x'],df['y'])
    ax1.set_xlabel('机器人四只的类别')
    ax1.set_ylabel('类别相对频率')g

    cumsum = np.cumsum(df['y'].values / np.sum(y))
    print(cumsum)
    ax2 = ax1.twinx()
    ax2.plot(x, cumsum, '-ro')
    
    pct_format = '{0:.0%}'
    formatted_weights = [pct_format.format(x) for x in cumsum]
    for i, txt in enumerate(formatted_weights):
        ax2.annotate(txt, (x[i], cumsum[i]), fontweight='heavy')  

x = np.array(["都没有", "都有", "只有腿", "只有轮子"])
y = np.array([15, 8, 63, 20])
pareto_plot(x, y)
```

```python
# 2.6

x = np.array(["Engineering & Design", "Procedures & Practices", "Management & Oversight", "Training & Communication"])
y = np.array([27, 24, 22, 10])

df = pd.DataFrame({
    'x': x,
    'y': y
})
df = df.sort_values(by='y', ascending=False)

fig, ax1 = plt.subplots(figsize=(20,10))
ax1.bar(df['x'],df['y'])
ax1.set_xlabel('Management System Cause Category')
ax1.set_ylabel('Number of Incidents')

cumsum = np.cumsum(df['y'].values / np.sum(y))
ax2 = ax1.twinx()
ax2.plot(x, cumsum, '-ro')

pct_format = '{0:.0%}'
formatted_weights = [pct_format.format(x) for x in cumsum]
for i, txt in enumerate(formatted_weights):
    ax2.annotate(txt, (x[i], cumsum[i]), fontweight='heavy')
```
